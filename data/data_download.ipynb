```python
from datasets import load_dataset, concatenate_datasets

# Load datasets from Hugging Face Hub
ds_24 = load_dataset("talzoomanzoo/rlmia_aime24", split="train")
ds_25 = load_dataset("talzoomanzoo/rlmia_aime25", split="train")

# Concatenate datasets
combined_dataset = concatenate_datasets([ds_24, ds_25])

# Check size
print(f"AIME24 size: {len(ds_24)}")
print(f"AIME25 size: {len(ds_25)}")
print(f"Combined size: {len(combined_dataset)}")

# Save locally in HuggingFace dataset format
combined_dataset.to_parquet("rlmia_aime24_25.parquet")
```
